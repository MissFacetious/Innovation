```python
import pandas as pd
import difflib
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('../data/foreign_investment.csv', skiprows=4)
```

```python
df.head()
```

```python
df.drop(columns=['Country Name', 'Indicator Name', 'Indicator Code'], inplace=True)
df.rename(columns={'Country Code': 'country_iso'}, inplace=True)
```

```python
df_new = df.melt(id_vars=["country_iso"], var_name="year", value_name="foreign_investment")
df_new
```

```python
df_new.dropna(inplace=True)
```

```python
df_new.to_csv('../data/converted/foreign_investment.csv', index=False)
```
