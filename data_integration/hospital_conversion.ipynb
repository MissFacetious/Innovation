```python
import pandas as pd
import difflib
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 100)
```

```python
hospitals = pd.read_csv('../data/hospitals.csv')
```

```python
hospitals.head()
```

```python
hospitals_new = hospitals.copy()
```

```python
countries = pd.read_csv('../data/country-codes_csv.csv')
```

```python
def get_country_iso(name):
    p = difflib.get_close_matches(name, countries['official_name_en'].dropna().unique(), n=1, cutoff=0.5)
    if len(p)>0:
        return countries[countries['official_name_en']==p[0]]['ISO3166-1-Alpha-3'].to_list()[0]
    else:
        return ""
```

```python
hospitals_new['country_iso'] = [get_country_iso(x) for x in hospitals_new['Country']]
hospitals_new[hospitals_new['country_iso']=='']
```

```python
hospitals_new.drop(columns=['Country'], inplace=True)
```

```python
hospitals_new.rename(columns={'Total density per 100 000 population: Hospitals': 'total_hospital_density_per_100k'}, inplace=True)
```

```python
hospitals_new.head()
```

```python
hospitals_new.drop(hospitals_new[hospitals_new.Year == 2010].index, inplace=True)
```

```python
hospitals_new
```

```python
hospitals_new.drop(82, inplace=True)
hospitals_new.drop(259, inplace=True)
```

```python
hospitals_new.to_csv('../data/converted/hospitals.csv', index=False)
```
