```python
import pandas as pd
import difflib
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('../data/mobile_subscriptions.csv', skiprows=4)
```

```python
df.head()
```

```python
df.drop(columns=['Country Name', 'Indicator Name', 'Indicator Code'], inplace=True)
```

```python
df.rename(columns={'Country Code': 'country_iso'}, inplace=True)
```

```python
df = df.melt(id_vars=["country_iso"], var_name="year", value_name="mobile_subscriptions")
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.to_csv('../data/converted/mobile_subscriptions.csv', index=False)
```
