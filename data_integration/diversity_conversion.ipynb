```python
import pandas as pd
import difflib
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('../data/diversity_2020.csv')
```

```python
df.head()
```

```python
import difflib
countries = pd.read_csv('../data/country-codes_csv.csv')
def get_country_iso(name):
    p = difflib.get_close_matches(name, countries['official_name_en'].dropna().unique(), n=1, cutoff=0.6)
    if len(p)>0:
        return countries[countries['official_name_en']==p[0]]['ISO3166-1-Alpha-3'].to_list()[0]
    else:
        return ""
```

```python
df['country_iso'] = [get_country_iso(x) for x in df['name']]
df[df['country_iso']=='']
```
